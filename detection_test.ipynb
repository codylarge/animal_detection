```python
from PytorchWildlife.models import detection as pw_detection
from PytorchWildlife import utils as pw_utils
import cv2
detection_model = pw_detection.MegaDetectorV6(version="MDV6-yolov10-c")
```

```python
primary_dir = "motion_videos"
secondary_dir = "event_2026-03-27_12-05AM"
filename = "motion_1774595150954_full.jpg"

path = f"{primary_dir}/{secondary_dir}/{filename}"

img = cv2.imread(path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

result = detection_model.single_image_detection(path)
pw_utils.save_detection_images(result, "test_output")
```
